```python
%matplotlib inline
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from node import Node, Graph
import numpy as np

# 生成 make_moons 数据集
X, y = make_moons(n_samples=2000, noise=0.1
                  , random_state=42)
# X是每个点的坐标(x, y)，y是每个点的label
nodes = []
index = 0
for df, label in zip(X, y):
    nodes.append(Node(df, label, index))
    index += 1
### 获得系数
from scipy.stats import pearsonr
from sklearn.neighbors import KernelDensity

# 计算相关系数
corr_coef, _ = pearsonr(X[:, 0], X[:, 1])
print("相关系数：", corr_coef.round(3))

# 绘制散点图
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("moon_dataset visulization")
plt.show()

# 核密度估计
kde = KernelDensity(kernel='gaussian', bandwidth=0.2).fit(X)
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))
Z = np.exp(kde.score_samples(np.c_[xx.ravel(), yy.ravel()])).reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap=plt.cm.Blues)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.title("Kernel Density Estimation")
plt.show()
```
