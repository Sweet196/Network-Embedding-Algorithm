```python
from sklearn.manifold import MDS

def mds2_2(X):
    # 使用MDS算法将数据映射到2维空间
    mds = MDS(n_components=2)
    X_mds = mds.fit_transform(X)
    return X_mds

def mds2_32_2(X):
    # 使用MDS算法将数据映射到2维空间
    mds1 = MDS(n_components=32)
    mds2 = MDS(n_components=2)
    X_mds1 = mds1.fit_transform(X)
    X_mds2 = mds2.fit_transform(X_mds1)
    return X_mds2
```

### 原始矩阵输入MDS得到嵌入

```python
from sklearn.datasets import make_blobs
from sklearn.manifold import MDS
from matplotlib import pyplot as plt

X, y = make_blobs(n_samples=2000, random_state=42, n_features=2,centers=8, cluster_std=0.8,
                  center_box=(-10.0,10.0))

# 可视化原始数据和MDS处理后的数据
plt.figure(figsize=(18, 5))

# 原始数据
plt.subplot(1, 3, 1)
for i in range(8):
    plt.scatter(X[y==i, 0], X[y==i, 1], label=f'Class {i}')
plt.title('Original Data')
plt.legend()

plt.title('Original Data')
plt.legend()
111
X_mds1 = mds2_2(X)
plt.subplot(1, 3, 2)
for i in range(8):
    plt.scatter(X[y==i, 0], X[y==i, 1], label=f'Class {i}')
plt.title('Original Data')
plt.legend()
plt.title('Dim 2->2 Embed')
plt.legend()

X_mds2 = mds2_32_2(X)
plt.subplot(1, 3, 3)
for i in range(8):
    plt.scatter(X[y==i, 0], X[y==i, 1], label=f'Class {i}')
plt.title('Original Data')
plt.legend()
plt.title('Dim 2->32->2 Embed')
plt.legend()

plt.show()
```
