```python
from sklearn.manifold import MDS

def mds2_2(X):
    # 使用MDS算法将数据映射到2维空间
    mds = MDS(n_components=2)
    X_mds = mds.fit_transform(X)
    return X_mds

def mds2_32_2(X):
    # 使用MDS算法将数据映射到2维空间
    mds1 = MDS(n_components=32)
    mds2 = MDS(n_components=2)
    X_mds1 = mds1.fit_transform(X)
    X_mds2 = mds2.fit_transform(X_mds1)
    return X_mds2
```

### 原始矩阵输入MDS得到嵌入

```python
from sklearn.datasets import make_moons
from sklearn.manifold import MDS
from matplotlib import pyplot as plt

# 生成make_moons数据集
X, y = make_moons(n_samples=2000, noise=0.1, random_state=42)

# 可视化原始数据和MDS处理后的数据
plt.figure(figsize=(18, 5))

# 原始数据
plt.subplot(1, 3, 1)
plt.scatter(X[y==0, 0], X[y==0, 1], color='r', label='Class 0')
plt.scatter(X[y==1, 0], X[y==1, 1], color='b', label='Class 1')
plt.title('Original Data')
plt.legend()
111
X_mds1 = mds2_2(X)
plt.subplot(1, 3, 2)
plt.scatter(X_mds1[y==0, 0], X_mds1[y==0, 1], color='r', label='Class 0')
plt.scatter(X_mds1[y==1, 0], X_mds1[y==1, 1], color='b', label='Class 1')
plt.title('Dim 2->2 Embed')
plt.legend()

X_mds2 = mds2_32_2(X)
plt.subplot(1, 3, 3)
plt.scatter(X_mds2[y==0, 0], X_mds2[y==0, 1], color='r', label='Class 0')
plt.scatter(X_mds2[y==1, 0], X_mds2[y==1, 1], color='b', label='Class 1')
plt.title('Dim 2->32->2 Embed')
plt.legend()

plt.show()
```

```python
from scipy.sparse.csgraph import dijkstra
from sklearn.datasets import make_moons
from sklearn.manifold import MDS
from sklearn.metrics import pairwise_distances
from matplotlib import pyplot as plt

# 生成make_moons数据集
X, y = make_moons(n_samples=2000, noise=0.1, random_state=42)
from sklearn.metrics import pairwise_distances

X = pairwise_distances(X)
X = dijkstra(X, directed=False)

print(X)

# 可视化原始数据和MDS处理后的数据
plt.figure(figsize=(18, 5))

# 原始数据
plt.subplot(1, 3, 1)
plt.scatter(X[y==0, 0], X[y==0, 1], color='r', label='Class 0')
plt.scatter(X[y==1, 0], X[y==1, 1], color='b', label='Class 1')
plt.title('Original Data')
plt.legend()

X_mds1 = mds2_2(X)
plt.subplot(1, 3, 2)
plt.scatter(X_mds1[y==0, 0], X_mds1[y==0, 1], color='r', label='Class 0')
plt.scatter(X_mds1[y==1, 0], X_mds1[y==1, 1], color='b', label='Class 1')
plt.title('Dim 2->2 Embed')
plt.legend()

X_mds2 = mds2_32_2(X)
plt.subplot(1, 3, 3)
plt.scatter(X_mds2[y==0, 0], X_mds2[y==0, 1], color='r', label='Class 0')
plt.scatter(X_mds2[y==1, 0], X_mds2[y==1, 1], color='b', label='Class 1')
plt.title('Dim 2->32->2 Embed')
plt.legend()

plt.show()
```
